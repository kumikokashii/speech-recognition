Currently, Bokeh graphs become invisible when a Jupyter notebook is uploaded to Github.

Below is a workaround to show Bokeh graphs on Github. It is done by temporarily creating a png image and displaying that image.

# Need this block in main

```python
import builtins  # Not needed unless saving for github
builtins.use_bokeh_screenshot = True  # Not needed unless saving for github
```

# Need this block in imported module with bokeh show

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook, export_png
output_notebook()
```

```python
import builtins
import IPython.display as ipd
import os

use_bokeh_screenshot = False
if hasattr(builtins, 'use_bokeh_screenshot'):
    use_bokeh_screenshot = builtins.use_bokeh_screenshot

if use_bokeh_screenshot:
    def show(p):
        filename = 'temp_img.png'
        ipd.display(ipd.Image(export_png(p, filename=filename)))
        os.remove(filename)
```

```python
# Sample plot
x = [0, 1, 2]
y = [12, 0, 20]

p = figure(title='Am I here?', x_axis_label='x', y_axis_label='y')
p.line(x, y, line_width=2)

show(p)
```
